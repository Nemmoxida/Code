```python
# Month - Per week expenses
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
```

```python
month = ["Jan", "Feb", "March", "April", "May", "June", "July", "Agust", "Sep", "Oct", "Nov", "Des"]

Week1 = np.random.randint(200000, 500001, 12)
Week2 = np.random.randint(200000, 500001, 12)
Week3 = np.random.randint(200000, 500001, 12)
Week4 = np.random.randint(200000, 500001, 12)

data = {
    "Month" : month,
    "Week 1" : Week1,
    "Week 2" : Week2,
    "Week 3" : Week3,
    "Week 4" : Week4
}

df = pd.DataFrame(data)

# print(df)
```

```python
# Table
df["Total Pengeluaran"] = df[["Week 1", "Week 2", "Week 3", "Week 4"]].sum(axis = 1)
df
```

```python
# Bar Chart

plot.figure(figsize=(12,5)) # 12 is Width and 5 is height
plot.bar(df["Month"], df["Total Pengeluaran"], color = "#ff8fe7")
plot.title("Expenses from this year")
plot.ylabel("Rupiah")
plot.xticks(rotation = -45)
plot.grid(axis="y")
# plot.tight_layout()
plot.show()
```

```python
plot.figure(figsize=(12,6))

for i in ["Week 1", "Week 2", "Week 3", "Week 4"]:
    plot.plot(df["Month"], df[i], marker="o", label=i)

plot.title("Expenses from this year")
plot.xlabel("Month")
plot.ylabel("Rupiah")
plot.grid(True)
plot.xticks(rotation = -45)
plot.legend()
plot.show()
```

```python
# Pie chart

plot.figure(figsize=(6,6))
plot.pie(df["Total Pengeluaran"], labels=df["Month"], startangle=90, autopct="%1.1f%%")
plot.title("Total Expenses from this year")
plot.axis("equal")
plot.show()
```

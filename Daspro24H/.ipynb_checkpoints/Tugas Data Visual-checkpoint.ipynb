```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression as LR
```

```python
df = pd.read_csv("kecelakaan.csv")
df = df.round(0)
```

```python
df["2016"] = df["2016"].round(0).astype(int)
df = df.fillna(0)
df
```

```python

df["2016"] = df["2016"].round(0).astype(int)
df["2017"] = df["2017"].round(0).astype(int)
df["2018"] = df["2018"].round(0).astype(int)
df["2019"] = df["2019"].round(0).astype(int)
df["2020"] = df["2020"].round(0).astype(int)
df[1:6]
```

```python
df["Total Kecelakaan"] = df[["2016", "2017", "2018", "2019", "2020"]].sum(axis = 1)
```

```python
df = df.drop("Satuan", axis=1)
```

```python
df[1:5]
```

```python
plot.figure(figsize=(12,5)) # 12 is Width and 5 is height
plot.bar(df["Daftar Data"][1:5], df["Total Kecelakaan"][1:5], color = "#ff8fe7")
plot.title("Statistik Kecelakaan Tahun 2016 - 2020 di Provinsi Kalimantan Timur")
plot.ylabel("Jumlah")
plot.xticks(rotation = -45)
plot.grid(axis="y")
plot.tight_layout()
plot.show()
```

```python
plot.figure(figsize=(6,6))
plot.pie(df["Total Kecelakaan"][1:5], labels=df["Daftar Data"][1:5], startangle=90, autopct="%1.1f%%")
plot.title("Statistik Kecelakaan")
plot.axis("equal")
plot.show()
```
